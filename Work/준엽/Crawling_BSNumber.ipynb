# 사업자등록번호 가져오기

```python
import pandas as pd
import csv
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time
import random

pd.options.display.max_columns = 1000
pd.options.display.max_rows = 1000
# row 생략 없이 출력
pd.set_option('display.max_rows', 1000)
# col 생략 없이 출력
pd.set_option('display.max_columns', 1000)
```

```python
#pd.read_csv가 사용이 안돼서 만든 def
def read_csv1(file_name):
    f = open(file_name,encoding='utf-8-sig')
    reader = csv.reader(f)
    csv_list=[]
    f.close

    for i in reader:
        csv_list.append(i)
    df = pd.DataFrame(csv_list)
    df = df.rename(columns = df.iloc[0])
    df = df.drop([0])
    return df
```

```python
df = read_csv1('main_df.csv')
```

```python
df.columns
```

```python
#기업명만 불러오기
df1 = df[df['사업자등록번호']==''][['사업자등록번호','기업명']]
df_b = df1['기업명'].unique().tolist()
df_b
```

```python
df_b1 = pd.DataFrame(df_b, columns=['기업명'])
```

```python
options = webdriver.ChromeOptions() 
options.add_argument("start-maximized")
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)
```

```python
#프록시 ip리스트
ipList=["112.217.199.122:55872","121.139.218.165:31409","183.111.26.15:8080","211.203.234.141:8080","222.121.116.26:47314"
    ,"115.21.163.39:8888","1.215.70.130:44072","210.107.78.173:3128","112.168.11.170:808","222.121.116.26:55149","175.125.216.117:8080"
    ,"211.244.224.130:8080","106.10.55.212:3128","119.192.195.83:8080","213.109.133.156:8080","23.237.173.102:3128","216.198.188.26:51068"
    ,"209.212.33.99:8080","64.20.33.202:8080","35.245.208.185:3128"
    ] 
```

```python
df_b2 = []
df_b2_name = []
```

```python
#크롤링 시작
count = 0

for i in range(len(df_b2)+1,len(df_b)):
    time.sleep(5)
    for i in ipList:
        try:
            webdriver.DesiredCapabilities.CHROME['proxy'] = {
                "httpProxy": i,
                "ftpProxy": i,
                "sslProxy": i,
                "proxyType": "MANUAL"
            }
            webdriver.DesiredCapabilities.CHROME['acceptSsICerts'] = True
            driver = webdriver.Chrome("./chromedriver.exe",chrome_options=options)
            driver.get("https://bizno.net/%EC%82%AC%EC%97%85%EC%9E%90%EB%93%B1%EB%A1%9D%EB%B2%88%ED%98%B8%EC%A1%B0%ED%9A%8C/")
            driver.implicitly_wait(2)
        except:
            continue

    elem = driver.find_element(By.XPATH, '//*[@id="query"]')
    time.sleep(4)
    elem.clear()
    time.sleep(6)
    elem.send_keys(df_b[i])
    print('a')

    elem = driver.find_element(By.XPATH, '//*[@id="home"]/div[2]/div/div/form/div/div[2]/button')
    time.sleep(1)
    elem.send_keys(Keys.ENTER)
    print('b')

    elem = driver.find_element(By.XPATH, '//*[@id="home"]/div[2]/div/div/h2[2]/strong/span')
    time.sleep(2)
    text = elem.get_attribute('innerText')
    print('c')

    if int(text) > 0:
        elem = driver.find_element(By.XPATH, '/html/body/section[2]/div/div/div[1]/div[1]/div/div/div/a')
        time.sleep(5)
        elem.send_keys(Keys.ENTER)
        print('d')
        time.sleep(6)
        elem1 = driver.find_element(By.XPATH, '/html/body/section[2]/div/div/div[1]/div[1]/div/div/div/a/h1')
        name = elem1.get_attribute('innerText')
        
        
        for i in range(1,20):
                elem = driver.find_element(By.XPATH, f'/html/body/section[2]/div/div/div[1]/div[1]/div/table/tbody/tr[{i}]')
                text = elem.get_attribute('innerText')
                if text.split('\t')[0] == '사업자등록번호':
                        bs = text.split('\t')[1]
                        print('e')
                        break
                else:
                    continue
    else:
        bs = ''
        name = ''
                
    driver.back()
    count += 1

    df_b2.append(bs)
    df_b2_name.append(name)
    print(count)
```
